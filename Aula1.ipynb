```python
df_chess.groupby(["victory_status"]).mean(numeric_only=True).plot(kind="pie"
                                                ,y="turns"
                                                ,autopct="%1.0f%%")
plt.title("Mdia de status de vitria")
```

```python
df_diabetes = pd.read_csv('/content/diabetes.csv',sep=',')
```

```python
df_diabetes.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_diabetes.drop(["Class variable"],axis=1)
y = df_diabetes["Class variable"]                 #ELe seria o target, para classificar
```

```python
#SERPARA OS DADOS o Train Test Split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, train_size=0.75)
```

```python
df_diabetes.shape
```

```python
#Algoritimo que classifica dados
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X=x_train,y=y_train)
```

```python
acurrence = knn.score(x_test, y_test)
```
